### Notes on the ID column

- The ID for the row. Wondering what kind of ID is it. It's a six digit number with a dash before the 3rd. Thought it was a date of some sort but that isn't right because they don't follow any of the typical date formats. There is a pattern emerging though. Some numbers repeat more than others.
- The ID's seem to go from 12- to 18-
- They are objects, not formatted numbers, so strings. Maybe I should convert them,but I'm not doing any operations on them. So leave them as is. But I do want to split them up between the first part of the ID and the second.
- So the second half of the number is a count. If you do a value counts of the first half and groupby.max() of the second half you get the same numbers (for instance 17 has 1038 counts, and also its max value in the secon half is 1038). Basically there is meaning in the first half only. They are grouped.
- There are 5105 entries. So whatever this grouping is for the first half of the ID, it will help me to better understand how to impute the data sense none of it is missing.
- Its only necessary to split out the first half of the ID, the second half seems to not matter sense it is a count. 
- **Mystery solved**. The date time column and the ID correspond to each other. It's the year. The second half is the count.
